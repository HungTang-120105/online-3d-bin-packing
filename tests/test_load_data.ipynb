```python
# Test with bottom_left_back model

from src.heuristics.bottom_left_back import BTB
from src.core.box import Box
import numpy as np

import torch

pt_file_path = '..\data\cut_1.pt'

data = torch.load(pt_file_path, map_location=torch.device('cpu'))
n = len(data)
utilization = []

for i in range(n):
    model = BTB(binsize=(10,10,10))
    boxes = [ Box(*(box_size)) for box_size in data[i] ] # Tạo danh sách các hộp từ dữ liệu

    for box in boxes:
        model.place_box(box)

    utilization.append(model.utilization()) # Tính toán tỷ lệ sử dụng không gian
    print("Tỷ lệ sử dụng không gian của mô hình {}: {}".format(i, model.utilization())) # In ra tỷ lệ sử dụng không gian của mô hình


print(np.array(utilization).mean()) # Tính toán trung bình tỷ lệ sử dụng không gian

```
